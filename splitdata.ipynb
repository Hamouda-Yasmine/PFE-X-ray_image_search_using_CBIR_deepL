```python
import pandas as pd
csvtrain=pd.read_csv("/content/drive/MyDrive/IRMA/ImageCLEFmed2009_train_codes.02.csv",sep=';')
csvtrain=csvtrain.set_index("image_id",drop=False)
csvtest=pd.read_csv("/content/drive/MyDrive/IRMA/ImageCLEFmed2009_test_codes.03.csv",sep=';')
csvtest=csvtest.set_index("image_id",drop=False)
path1 = "/content/drive/MyDrive/clef2009/test/*.*"
```

```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import sys

labels = pd.read_csv(r'/content/drive/MyDrive/IRMA/ImageCLEFmed2009_train_codes.02.csv')

train_dir =r'/content/drive/MyDrive/IRMA/train'
DR = r"/content/drive/MyDrive/IRMA/datasplittrain/"
if not os.path.exists(DR):
    os.mkdir(DR)

for filename, class_name in labels.values:
    # Create subdirectory with `class_name`
    if not os.path.exists(DR + str(class_name)):
        print()
        os.mkdir(DR + str(class_name))
    src_path = train_dir + '/'+ filename + '.jpeg'
    dst_path = DR + str(class_name) + '/' + filename + '.jpeg'
    try:
        shutil.copy(src_path, dst_path)
        print("sucessful")
    except IOError as e:
        print('Unable to copy file {} to {}'
              .format(src_path, dst_path))
    except:
        print('When try copy file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```
